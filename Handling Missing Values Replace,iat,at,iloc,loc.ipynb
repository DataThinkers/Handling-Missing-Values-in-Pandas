```python
import pandas as pd
```

```python
data = pd.read_csv('data_#na.csv')
data
```

####  Replace  "Priyang" with "Anil"

```python
#data.replace(to_replace='Priyang',value='Anil')
data.replace('Priyang','Anil')
```

#### Replace Name "Priyang" and "Aadhya" with "Anil"

```python
data.replace(['Priyang','Aadhya'],'Anil')
```

#### Replace "NaN" with 40

```python
import numpy as np
data.replace(np.nan,40)
```

#### Replace Marks column NaN with 40 and Grades column NaN with FF

```python
data.replace({'Marks':{np.nan:40},
             'Grades':{np.nan:'FF'}})
```

```python
data
```

#### Replace 'not available',np.nan,'-99' of Grades column to 'FF' 

```python
data.replace({'Grades':['not available',np.nan,'-99']},'FF',
            inplace=True)
```

```python
data
```

##### Replace Grades values 
#### FF : Fail
#### AB : Very Good
#### AA : Excellent
#### BA : Good

```python
data.replace(['FF','AB','AA','BA'],
             ['Fail','Very Good','Excellent','Good'])
```

#### Replace NaN values with ffill

```python
data.replace(np.nan,method='ffill',limit=1)
```

#### Replace  consecutive NaN values of Grades column with ''FF" Using iloc

```python
data=pd.read_csv('data_#na.csv')
data.iloc[3:5,2]='FF'
data
```

#### Replace consecutive NaN values of Grades column with ''FF" Using loc

```python
data=pd.read_csv('data_#na.csv')
data.loc[3:4,'Grades']='FF'
data
```
